**Description of individual Variable**

1. **hotel** : *Hotel(Resort Hotel or City Hotel)*

2. **is_canceled** : *Value indicating if the booking was canceled (1) or not (0)*

3. **lead_time** :* Number of days that elapsed between the entering date of the booking into the PMS and the arrival date*

4. **arrival_date_year** : *Year of arrival date*

5. **arrival_date_month** : *Month of arrival date*

6. **arrival_date_week_number** : *Week number of year for arrival date*

7. **arrival_date_day_of_month** : *Day of arrival date*

8. **stays_in_weekend_nights** : *Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel*

9. **stays_in_week_nights** : *Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel*

10. **adults** : *Number of adults*

11. **children** : *Number of children*

12. **babies** : *Number of babies*

13. **meal** : *Type of meal booked. Categories are presented in standard hospitality meal packages:*

14. **country** : *Country of origin.`*

15. **market_segment** : *Market segment designation. In categories, the term “TA” means “Travel Agents” and “TO” means “Tour Operators”*

16. **distribution_channel** : *Booking distribution channel. The term “TA” means “Travel Agents” and “TO” means “Tour Operators”*

17. **is_repeated_guest** : *Value indicating if the booking name was from a repeated guest (1) or not (0)*

18. **previous_cancellations** : *Number of previous bookings that were cancelled by the customer prior to the current booking*

19. **previous_bookings_not_canceled** : *Number of previous bookings not cancelled by the customer prior to the current booking*

20. **reserved_room_type** : *Code of room type reserved. Code is presented instead of designation for anonymity reasons.*

21. **assigned_room_type** : *Code for the type of room assigned to the booking.*

22. **booking_changes** : *Number of changes/amendments made to the booking from the moment the booking was entered on the PMS until the moment of check-in or cancellation*

23. **deposit_type** : *Indication on if the customer made a deposit to guarantee the booking.*

24. **agent** : *ID of the travel agency that made the booking*

25. **company** : *ID of the company/entity that made the booking or responsible for paying the booking.*

26. **days_in_waiting_list** : *Number of days the booking was in the waiting list before it was confirmed to the customer*

27. **customer_type** : *Type of booking, assuming one of four categories*


28. **adr** : *Average Daily Rate as defined by dividing the sum of all lodging transactions by the total number of staying nights*

29. **required_car_parking_spaces** : *Number of car parking spaces required by the customer*

30. **total_of_special_requests** :* Number of special requests made by the customer (e.g. twin bed or high floor)*

31. **reservation_status** : *Reservation last status, assuming one of three categories*
* Canceled – booking was canceled by the customer
* Check-Out – customer has checked in but already departed
* No-Show – customer did not check-in and did inform the hotel of the reason why





32. **reservation_status_date** : *Date at which the last status was set. This variable can be used in conjunction with the ReservationStatus to understand when was the booking canceled or when did the customer checked-out of the hotel*

## Data Cleaning

```python
# Finding total rows
Total_num_of_rows=df1.shape[0]
print(Total_num_of_rows)
```

```python
# checking number of duplicate rows
df1.duplicated().value_counts()   #true means duplicate rows , false means non-duplicate rows
```

```python
# Visulizing through Count pot
plt.figure(figsize=(6,4))
sns.countplot(x=df1.duplicated())      # Green (True) are duplicates , needs to be deleted
```

So we have 31994 duplicate rows in our data. So we will drop the duplicate rows from our data

```python
#dropping the duplicate rows
df1= df1.drop_duplicates()
```

```python
# data set reduced
df1.shape
```

```python
#to fill the NaN value in the columns, let's check which colomns has null value
#adding all Null values column-wise and sorting in descending order
#checking Null values till 6th result only becoz only 4 columns has Null values , for display ussing reset index function
miss_values=df1.isna().sum().sort_values(ascending=False)
print(miss_values)
```

```python
# Visulaizing null values through heatmap.
plt.figure(figsize=(25, 10))
sns.heatmap(df1.isnull(), cbar=False, yticklabels=False,cmap='viridis')
plt.xlabel("Name Of Columns")
plt.title("Places of missing values in column")
```

```python
#lets check the percentage of null value in each column
#starting from column named "company"

percentage_company_null = (miss_values[0] / uniq_num_of_rows)*100
print(round(percentage_company_null, ndigits=2))         # rounding upto 2 decimal place and printing %
```

```python
# It is better to drop the column 'company' since the number of missing values is extremely high compared to the number of rows.

df1.drop(labels=['company'], axis=1, inplace=True)
```

```python
# now let's check for agent

percentage_agent_null = (miss_values[1] / uniq_num_of_rows)*100
print(round(percentage_agent_null, ndigits=2))
```

```python
# As we have seen, there is minimul null values in agent, Lets fill these value by taking mode of the all values

df1['agent'].fillna(value = 0, inplace = True)
df1['agent'].isnull().sum() # we re-check that column has no null value
```

```python
#Check the percentage null value in country col

percentage_country_null = (miss_values[2] / uniq_num_of_rows)*100
print(round(percentage_country_null, ndigits=2))
```

```python
# We have less null vlues in country column, so we will replace null with 'other' as country name.

df1['country'].fillna(value = 'others', inplace = True)
df1['country'].isnull().sum() # we re-check that column has no null value
```

```python
#Check the percentage of null value in children column

percentage_children_null = (miss_values[3] / uniq_num_of_rows)*100
print(round(percentage_children_null, ndigits=2))
```

```python
# We have almost nil Null Values in children column, so we will replace null with 0

df1['children'].fillna(value = 0, inplace = True)
df1['children'].isnull().sum() # we re-check that column has no null value
```

```python
# Visulaizing null values through heatmap (after Null values are removed)
plt.figure(figsize=(20, 10))
sns.heatmap(df1.isnull(), cbar=False, yticklabels=False,cmap='viridis')
plt.xlabel("Name Of Columns")
plt.title("Places of missing values in column")
```

##Addition of new column as per requirement

```python
#total number of people
df1['total_people'] = df1['adults'] + df1['babies'] + df1['children']
```

```python
#total stay in nights
df1['total_stay_in_nights'] = df1['stays_in_week_nights'] + df1['stays_in_weekend_nights']
df1['total_stay_in_nights'] # We have created a column for total stays in nights by adding week night & weekend nights stay col.
```

```python
# We have created a column for revenue using total stay * adr
df1['revenue'] = df1['total_stay_in_nights'] *df1['adr']
df1['revenue']
```

```python
# for our easy understanding, in column 'is_canceled': we will replace the value from (0,1) to not_cancelled, is cancelled.

df1['is_canceled'] = df1['is_canceled'].replace([0,1], ['not cancelled', 'is cancelled'])
df1['is_canceled']
```

```python
#Same for 'is_repeated_guest' column
df1['is_repeated_guest'] = df1['is_repeated_guest'].replace([0,1], ['not repeated', 'repeated'])
df1['is_repeated_guest']
```

```python
#Now, we will check overall revenue hotel wise
hotel_wise_total_revenue = df1.groupby('hotel')['revenue'].sum()
hotel_wise_total_revenue
```

#10) Does the guests alloted with the same room type which was reserved by them?

```python
# Define a function for the reserved room type is same  or not as  assigned room type
def same_room_allotment(x):
  '''function to check whether the reserved room
  type is same  or not as  assigned room type'''

  if x['reserved_room_type'] != x['assigned_room_type']:
    return 1
  else:
    return 0
```

```python
# create new column to df1
df1['Same_room_alloted_or_not']=df1.apply(lambda x: same_room_allotment(x),axis=1)

#groupby is canceled
grp_by_canceled_df=df1.groupby('is_canceled')

# create DF and calculate  percentage of same room alloted  or not
DF=pd.DataFrame(grp_by_canceled_df['Same_room_alloted_or_not'].sum()*100/grp_by_canceled_df.size()).rename(columns={0:"percentage"})
DF

#set plot size and plot barchart
plt.figure(figsize=(8,4))
sns.barplot(x=DF.index,y=DF['percentage'])

# set labels
plt.title('%  of not getting the same room vs Booking cancellation')
plt.xlabel('is_canceled')
plt.ylabel('% of not getting same room as per reservation')
plt.xticks([0,1],['Not canceled','canceled'])
```

### **Obseravtion:**


*   Its is clear that there is not much(2.5%) effect on cancellation of the bookings even if the guests are not assigned with rooms which they reserved during booking.
